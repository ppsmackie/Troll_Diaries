```python
user_times = {}
creation_dates = {}

import time
t0 = time.time()

### Create dictionary of users and list of tweet times ###
with open(tweets_dest, encoding="utf8") as f:
    reader = csv.reader(f,dialect='excel')
    current = next(reader)
    print(build_row(current,[1,13,9]))
    for i in range(total_tweets-1):
        current = next(reader)
        user = current[1]
        t = current[13]
        if user in user_times:
            user_times[user].append(t)
        else: 
            user_times[user] = [t] 
            creation_dates[user] = current[9]
f.close()

t1 = time.time()
```

```python
### Parse user time strings into datetime objects and store ordered list of each users tweet times

t2 = time.time()

user_datetimes = {}
for user in user_times:
    dates = user_times[user].copy()
    datetimelist = []
    for j in range(len(dates)):
        datetime_tweet = datetime.strptime(dates[j], '%Y-%m-%d %H:%M')
        datetimelist.append(datetime_tweet)
    datetimelist.sort()
    user_datetimes[user] = datetimelist
```

```python
### Check all tweets included
T = 0
for user in user_datetimes:
     T += len(user_datetimes[user])
print(T)
```

```python
### Create list of each users tweet dates

t2_5 = time.time()

user_dates = {}
todate = lambda x: x.date()
for user in user_datetimes:
    datetimes = user_datetimes[user]
    dates = list(map(todate, datetimes))
    user_dates[user] = dates
```

```python
### Test all tweets included
T = 0
for user in user_datetimes:
     T += len(user_dates[user])
print(T)
```

```python
### find first and last tweet days and calculate days between them for length of vector
firsttweet = datetime.now().date()
lasttweet = datetime(2000,1,1).date()

for user in user_dates:
    dates = user_dates[user]
    first = dates[0]
    last = dates[-1]
    if first < firsttweet: firsttweet = first
    if last > lasttweet: lasttweet = last
print('First tweet: ', firsttweet)
print('Last tweet: ', lasttweet)
total_days = (lasttweet - firsttweet).days + 1
print('Days spanned: ', total_days)
print('No. Users: ', len(keys))
```

```python
### create and save a dictionary keyed by username of vectors with number of tweets per day
date_vectors = {}
for user in user_dates:
    day_vector = np.zeros(total_days, int)
    dates = user_dates[user]
    for date in dates:
        date_obj = date
        day_number = (date_obj - firsttweet).days
        day_vector[day_number] += 1
    date_vectors[user] = day_vector
```

```python
# Create data matrix (number of tweets each user authored each day)
data_matrix = np.zeros((total_users,total_days),int)
i = 0
for user in date_vectors:
    vect = date_vectors[user]
    data_matrix[i] = vect
    i += 1
print(np.sum(data_matrix))
#pick(data_matrix,'data_matrix')
```

```python
### convert date object to array index

def dtn(date):
    day1 = datetime(2009,5,9).date()
    return (date - day1).days

day2 = datetime(2009,5,10).date()
election = datetime(2016,11,8).date()

print(dtn(day2))
print(dtn(election))
```

```python
### Create a matrix of account creation date, first tweet and last tweet

matrix = np.zeros((total_users,3),int)
i = 0
for user in creation_dates:
    cd = creation_dates[user]
    cd = datetime.strptime(cd, '%Y-%m-%d').date()
    cd = dtn(cd)
    if cd < 0: cd = 0
    dates = user_dates[user]
    fd = dates[0]
    ld = dates[-1]
    fd = dtn(fd)
    ld = dtn(ld)
    matrix[i] = [cd,fd,ld]
    i += 1
```

```python
### Plot the total number of users active over the period spanned
### Note that most accounts had a long resting period before activation
cds = matrix[:,0]
fds = matrix[:,1]
lds = matrix[:,2]

cr = np.zeros(3331)
ac = np.zeros(3331)
for i in range(3667):
    c = cds[i]
    f = fds[i]
    l = lds[i]
    cr[c:l] += 1
    ac[f:l] += 1
    
plt.plot(cr)
plt.plot(ac)
plt.show()
```
